```python
# split the 'Name' column into company and model
def split_name(name_str):
    two_word_companies = ['Land Rover', 'Mercedes Benz', 'Maruti Suzuki', 'Mercedes-Benz', 'Rolls Royce', 'Mini Cooper', 'MINI Cooper']
    name_parts = name_str.split()

    if ' '.join(name_parts[:2]) in two_word_companies:
        company_name = ' '.join(name_parts[:2])
        model_name = ' '.join(name_parts[2:])
    else:
        company_name = name_parts[0]
        model_name = ' '.join(name_parts[1:])

    return pd.Series([company_name, model_name])

df2[['Company', 'Model']] = df2['car_name'].apply(split_name)
df2[['Company', 'Model']]
```

## car_price_in_rupees

```python
df2 = df2.rename(columns={'car_price_in_rupees': 'car_price'})
```

```python
df2['car_price'] = df2['car_price'].str.replace('₹', '', regex=False)
df2['car_price'] = df2['car_price'].str.strip() # remove leading/trailing spaces
```

```python
df2 = df2.sort_values(by=['car_price'], ascending=False)
```

```python
def convert_car_price(price_str):
    price_str = str(price_str)
    price_str = price_str.replace(',', '')

    if 'Lakh' in price_str:
        price_str = price_str.replace(' Lakh', '')
        price = float(price_str) * 100000
    elif 'Crore' in price_str:
        price_str = price_str.replace(' Crore', '')
        price = float(price_str) * 10000000
    else:
        price = float(price_str)

    return price
```

```python
df2['car_price'] = df2['car_price'].apply(convert_car_price)
```
