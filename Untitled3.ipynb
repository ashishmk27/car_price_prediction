```python
import pandas as pd

def split_car_name(name):
  """Splits a car name into company and model.

  Args:
    name: The car name string.

  Returns:
    A pandas Series containing the company and model.
  """
  two_word_companies = ['Land Rover', 'Mercedes Benz', 'Maruti Suzuki', 'Mercedes-Benz', 'Rolls Royce', 'Mini Cooper', 'MINI Cooper']
  name_parts = name.split()

  if ' '.join(name_parts[:2]) in two_word_companies:
    company = ' '.join(name_parts[:2])
    model = ' '.join(name_parts[2:])
  else:
    company = name_parts[0]
    model = ' '.join(name_parts[1:])

  return pd.Series([company, model], index=['company', 'model'])

# Apply the function to the 'car_name' column and create new columns
df[['company', 'model']] = df['car_name'].apply(split_car_name)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd

# Function to clean and convert prices
def clean_price(price_str):
    """Removes 'Lakh' and converts to numeric, multiplying by 100000."""
    if isinstance(price_str, str) and 'Lakh' in price_str:  # Check if it's a string and contains 'Lakh'
        price_str = price_str.replace('₹', '').replace('Lakh', '').strip()  # Remove ₹ and Lakh
        try:
            return float(price_str) * 100000  # Convert to float and multiply
        except ValueError:
            print(f"Warning: Could not convert '{price_str}' to a number.")
            return None  # Handle cases where conversion fails
    elif isinstance(price_str, str) and 'Crore' in price_str:
         price_str = price_str.replace('₹', '').replace('Crore', '').strip()  # Remove ₹ and Crore
         try:
             return float(price_str) * 10000000
         except ValueError:
             print(f"Warning: Could not convert '{price_str}' to a number.")
             return None  # Handle cases where conversion fails
    else:
        return price_str  # Keep as is if not a string or doesn't contain 'Lakh'

# Apply the function to the 'car_price_in_rupees' column
df['car_price_in_rupees'] = df['car_price_in_rupees'].apply(clean_price)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd

def clean_kms_driven(kms_driven_str):
    """Removes 'km' and commas, and converts to numeric."""
    if isinstance(kms_driven_str, str):
        kms_driven_str = kms_driven_str.replace('km', '').replace(',', '').strip()
        try:
            return int(kms_driven_str)  # Convert to integer
        except ValueError:
            print(f"Warning: Could not convert '{kms_driven_str}' to a number.")
            return None  # Handle cases where conversion fails
    else:
        return kms_driven_str  # Keep as is if not a string

# Apply the function to the 'kms_driven' column
df['kms_driven'] = df['kms_driven'].apply(clean_kms_driven)

# Display the updated DataFrame
print(df)
```
